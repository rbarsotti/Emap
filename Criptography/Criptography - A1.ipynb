```python
# Algoritmo Euclidiano Estendido

def exeuclids(a, b):
    x=1
    x1=0
    y=0
    y1=1
    while b != 0:
        q=int(a/b)
        rt = b
        xt=x1
        yt=y1
        b = a - b*q
        x1 = x - x1*q
        y1 = y - y1*q
        a=rt
        y=yt
        x=xt
    return (a, x, y)

def solver(a, b, c):
    d = exeuclids(a, b)[0]
    if c%d !=0:
        return 0
    else:
        a = a/d
        b = b/d
        c = c/d
        a1 = exeuclids(a,b)[1]
        b1 = exeuclids(a,b)[2]
        a1 = c*a1
        b1 = c*b1
        print("{}{} + {}{} = {}".format(a,a1,b,b1,c))

solver(10,3,8)    
```

```python
# MDC
import random 
import math as m
aprox = 6/(m.pi**2)

# Euclids
def mdc(a,b):
    r0 = a % b
    if r0 == 0:
        return b
    r1 = b % r0
    if r1 == 0:
        return r0
    while r1 > 0:
        r = r0 % r1
        r0 = r1
        r1 = r
    return r0
        

def randomTuple():
    a = random.randint(0,100)
    b = a + 1
    while b > a or b == 0:
        b = random.randint(0,100)
    return (a,b)


def quociente(n):
    for i in range(n):
        ab = randomTuple()
        a = ab[0]
        b = ab[1]
        md = mdc(a,b)
        mdc_1 = 0
        if md == 1:
            mdc_1 += 1
        return mdc_1 / n

def simulate():
    l = [10,10**2,10**3,10**4,10**5]
    for i in l:
        for j in range(10):
            q_i = quociente(i)
            print(i,l,q_i)

simulate()
```

```python
# Fermat

def fermat(n):
    x = int(m.sqrt(n))
    if n == x**2:
        return (x, n)
    else:
        while True:
            x += 1
            y = m.sqrt((x**2)-n)
            if x == (n+1)/2.:
                return x
            if y == int(y):
                return (int(x+y),int(x-y))
    

fermat(10)
```

###### 3.12)

```python
def pi_1(x):
    if x % 4 == 1:
        return True
    
def pi_3(x):
    if x % 4 == 3:
        return True
    
def crivo(n):
    s_p = []
    sieve = [1] * n
    for p in range(2, n):
        if sieve[p] == 1:
            s_p.append(p)
            for i in range(p*p, n, p):
                sieve[i] = False
    return s_p

def pi_1n(x):
    p1 = 0
    for i in x:
        if pi_1(i) == True:
            p1 += 1
    return p1

def pi_3n(x):
    p3 = 0
    for i in x:
        if pi_3(i) == True:
            p3 += 1
    return p3
    

def fraction():
    for i in range(100,10**5,100):
        j = crivo(i)
        a = pi_1n(j)
        b = pi_3n(j)
        c = a / b
        print(a,b,c)
        
fraction()
```

```python
def powerN(a,k,n):
    cont = 1
    while k != 0:
        if k%2 == 0 :
            k = k/2
        else:
            cont = (a*cont)%n
            k = (k - 1)/2
        a = (a**2)%n
    return cont

powerN(10,133,7)
```

```python
def isMiller(n, b):
    a=1
    c = (n-1)/2
    while True:
        if c%2 != 0:
            break
        c = c/2
        a += 1
    i=0
    r = (b**c)%n
    while True:
        if (i ==0 and  r == 1) or (i>=0 and r == n-1):
            return False
        i += 1
        r = (r**2)%n
        if i>=a:
            return True
```
