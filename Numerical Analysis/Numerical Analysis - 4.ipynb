| y | 2 |3 | 5|

$

$   
S(x) = 
     \begin{cases}

       S_{1}(s) = ex^3 + fx^2 + gx + h  \;\;\;x \in [2,3] \\       
     \end{cases}
$

$\\
$

$S(1) = S_{0}(1) = a + b + c + d = 2$

$S(2) = S_{0}(2) = 8a + 4b + 2c + d = 3$

$
S'(x) =

#### 2) Sem auxílio de computador, mostre detalhadamente como obter um spline quadrático $S(x)$ talque existe $S(x)$ para todo $x \in \mathbb{R}$ e $S(x)$ interpola os pontos $(−9, 2), (−5, 2), (−1, 2), (2, 2), (7, 2),(10, 2)$ e $(32, 2)$.


| x | -9|-5|-1|2|7|10|32|
|---|---|--|--|-|-|--|--|
| y | 2 |2 | 2|2|2|2 | 2|

$\\
$


$   
S(x) = 
     \begin{cases}
       S_{0}(s) = ax^2 + bx + c  \;\;\;x \in [-9,-5] \\
       S_{1}(s) = dx^2 + ex + f  \;\;\;x \in [-5,-1] \\
       S_{2}(s) = gx^2 + hx + i  \;\;\;x \in [-1,2] \\
       S_{3}(s) = jx^2 + kx + l  \;\;\;x \in [2,7] \\
       S_{4}(s) = mx^2 + nx + o  \;\;\;x \in [7,10] \\
       S_{5}(s) = px^2 + qx + r  \;\;\;x \in [10,32] \\
     \end{cases}
$

$\\
$

$S(2) = S_{0}(-9) = 81a + -9b + c = 2$

$S(2) = S_{0}(-5) = 25a + -5b + c = 2$

$S(2) = S_{1}(-5) = 25d + -5e + f = 2$

$S(2) = S_{1}(-1) =   d +  -e + f = 2$

$S(2) = S_{2}(-1) =   g +  -h + i = 2$

$S(2) = S_{2}(2) =   4g +  2h + i = 2$

$S(2) = S_{3}(2) =   4j + 2k  + l = 2$

$S(2) = S_{3}(7) =  49j + 7k  + l = 2$

$S(2) = S_{4}(7) =  49m + 7n  + o = 2$

$S(2) = S_{4}(10) =100m +10n  + o = 2$

$S(2) = S_{5}(10) =100p +10q  + r= 2$

$S(2) = S_{5}(32)=1024p +32q  + r= 2$
