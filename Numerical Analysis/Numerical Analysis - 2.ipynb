# Numerical Analysis

#### Seja $f(x) = \frac{1}{1+x^2}$. Denote por $p_n(x)$ o polinomio interpolador de $f(x)$ de grau menor ou igual a $n$ obtido a partir dos $n+1$ pontos que dividem o intervalo $[−5, 5]$ em intervalos de mesmo tamanho.

##### (a) Obtenha $p_n(x)$ na forma de Newton para $n = 2$ e $n = 4$.

$p_2(x) = 0.03846154 + 0.19230769(x + 5) - 0.03846154(x + 5)x$

$p_4(x) = 0.03846154 + 0.0397878(x + 5) + 0.06100796(x + 5)(x + 2.5) - 0.0265252(x + 5)(x + 2.5)x + 0.00530504(x + 5)(x + 2.5)x(x - 2.5)$ 

```python
import numpy as np
import matplotlib.pyplot as plt

def x_p(n):
    x = (5 - (-5))/n
    xpt = []
    for i in range(n+1):
        xpt.append(i*x-5)
    return np.array(xpt)

def y_p(x):
    ypt = []
    for i in x:
        ypt.append(1/(1+i**2))

xpt2 = x_p(2)
xpt4 = x_p(4)
xpt8 = x_p(8)
xpt64 = x_p(64)

ypt2 = y_p(xpt2)
ypt4 = y_p(xpt4)
ypt8 = y_p(xpt8)
ypt64 = y_p(xpt64)

def NDD(x,y):
    n = len(x)
    A = np.zeros((n,n+1))
    A[:,0]= x
    A[:,1]= y
    for j in range(2,n+1):
        for i in range(j-1,n):
            A[i,j] = (A[i,j-1]-A[i-1,j-1]) / (A[i,0]-A[i-j+1,0])
    p = np.zeros(n)
    for k in range(0,n):
        p[k] = A[k,k+1]
    return p

def poly(t,x,p):
    n = len(x)
    out = p[n-1]
    for i in range(n-2,-1,-1):
        out = out*(t-x[i]) + p[i]
    return out
```
